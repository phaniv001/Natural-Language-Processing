```python
import numpy as np
import pandas as pd
from flair.embeddings import WordEmbeddings
from flair.embeddings import TransformerWordEmbeddings
from flair.data import Sentence
from scipy.spatial import distance
```

```python
glove_embedding = WordEmbeddings('glove')
```

```python
sentence_1 = Sentence('apple release iphone 12 pro max in 2020')
```

```python
glove_embedding.embed(sentence_1)
```

```python
for token in sentence_1:
    print(token)
    print(token.embedding)
    print('\n')
```

```python
sentence_1[0]
```

```python
sentence_1[0].embedding
```

```python
sentence_2 = Sentence('an apple a day keep the doctor away')
glove_embedding.embed(sentence_2)
```

```python
sentence_2[1]
```

```python
sentence_2[1].embedding
```

```python
for token in sentence_2:
    print(token)
    print(token.embedding)
    print('\n')
```

```python
#Glove distance between the same the words.

distance.euclidean(sentence_1[0].embedding, sentence_2[1].embedding)
```

```python
bert_embedding = TransformerWordEmbeddings('bert-base-multilingual-cased')
```

```python
bert_embedding.embed(sentence_1)

for token in sentence_1:
    print(token)
    print(token.embedding)
    print("\n")
```

```python
bert_embedding.embed(sentence_2)
for token in sentence_2:
    print(token)
    print(token.embedding)
    print("\n")
```

```python
distance.euclidean(np.array(sentence_1[0].embedding), np.array(sentence_2[1].embedding))
```
